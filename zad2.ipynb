```python
from scipy import stats
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parametry rozkładu normalnego
mu = 3          # Średnia
sigma = 1       # Odchylenie standardowe
```

**Podejście z określaniem zakresu danych (od ... do ...)**

```python
# Generowanie zakresów x1 i x2 w przypadku gdy określamy zakres danych
x1 = np.linspace(-7, 13, 100)       # 100 punktów dla 100 próbek
x2 = np.linspace(-7, 13, 10000)     # 10 000 punktów dla 10 000 próbek

# Funkcja rozkładu normalnego (PDF)
def gauss(x, sigma=1, mu=3):
    return stats.norm.pdf(x, mu, sigma)

# Tworzenie wykresów obok siebie (1 rząd, 2 kolumny)
fig, axes = plt.subplots(1, 2, figsize=(14, 8))

# Wykres dla 100 próbek 
axes[0].plot(x1, gauss(x1, sigma, mu), label='100 próbek')
axes[0].set_title('100 próbek')
axes[0].legend()

# Wykres dla 10 000 próbek 
axes[1].plot(x2, gauss(x2, sigma, mu), label='10 000 próbek')
axes[1].set_title('10 000 próbek')
axes[1].legend()

# Wyświetlenie wykresów
plt.tight_layout()  # Zapewnia, że wykresy nie będą się nakładały
plt.show()
```

**Podejście z generowaniem danych losowych z rozkładu normalnego o zadanej średniej i odchyleniu standardowym**

```python
# Generowanie danych z rozkładu normalnego
data_100 = np.random.normal(mu, sigma, 100)    # 100 próbek
data_10000 = np.random.normal(mu, sigma, 10000)  # 10 000 próbek
```

```python
plt.figure(figsize=(10, 6))

# Histogram dla 100 próbek
plt.hist(data_100, bins=10, color='blue', edgecolor='black', alpha=0.7, density=True, label='100 próbek')

# Histogram dla 10 000 próbek
plt.hist(data_10000, bins=30, color='green', edgecolor='black', alpha=0.7, density=True, label='10 000 próbek')

# Tytuł i etykiety
plt.title('Histogramy dla 100 i 10 000 próbek z rozkładu normalnego')
plt.xlabel('Wartości')
plt.ylabel('Gęstość')

# Legenda
plt.legend()

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```

średnia jest taka sama = 3

```python
mean_100 = np.round(np.mean(data_100))
print(mean_100)
```

```python
mean_10000 = np.round(np.mean(data_10000))
print(mean_10000)
```

```python
q0_10000 = np.quantile(data_10000, 0.0)
q1_10000 = np.round(np.quantile(data_10000, 0.25))
q3_10000 = np.round(np.quantile(data_10000, 0.75))
q4_10000 = np.round(np.quantile(data_10000, 1.0))
print(f'Q0: {q0_10000}')
print(f'Q1: {q1_10000}')
print(f'Q2: {median_10000}')
print(f'Q3: {q3_10000}')
print(f'Q4: {q4_10000}')
```

```python
# Tworzenie wykresów
fig, axes = plt.subplots(1, 2, figsize=(14, 8))  # 1 wiersz, 2 kolumny

# Histogram i krzywa Gaussa dla 100 próbek
axes[0].hist(data_100, bins=10, color='blue', edgecolor='black', alpha=0.7, density=True)  # 'density=True' normalizuje histogram
x_values = np.linspace(min(data_100), max(data_100), 1000)  # Zakres x
gaussian_curve = norm.pdf(x_values, mu, sigma)  # Krzywa Gaussa
axes[0].plot(x_values, gaussian_curve, color='red', label='Krzywa Gaussa')
axes[0].set_title('Histogram dla 100 próbek z krzywą Gaussa')
axes[0].legend()

# Histogram i krzywa Gaussa dla 10 000 próbek
axes[1].hist(data_10000, bins=100, color='green', edgecolor='black', alpha=0.7, density=True)  # 'density=True' normalizuje histogram
x_values_10000 = np.linspace(min(data_10000), max(data_10000), 1000)  # Zakres x
gaussian_curve_10000 = norm.pdf(x_values_10000, mu, sigma)  # Krzywa Gaussa
axes[1].plot(x_values_10000, gaussian_curve_10000, color='red', label='Krzywa Gaussa')
axes[1].set_title('Histogram dla 10 000 próbek z krzywą Gaussa')
axes[1].legend()

# Ustalenie wspólnej skali y dla obu wykresów
# Obliczanie maksimum wartości gęstości prawdopodobieństwa (krzywa Gaussa)
max_density = max(max(gaussian_curve), max(gaussian_curve_10000))

# Ustawianie tej samej skali dla obu wykresów
axes[0].set_ylim(0, max_density * 1.1)
axes[1].set_ylim(0, max_density * 1.1)

# Dopasowanie układu wykresów, aby nie nachodziły na siebie
plt.tight_layout()

# Wyświetlenie wykresów
plt.show()
```

```python
standard_deviation_100 = np.std(data_100, ddof=1)
print(standard_deviation_100)
```

```python
standard_deviation_10000 = np.std(data_10000, ddof=1)
print(standard_deviation_10000)
```

**Odp.** W przypadku większej liczby próbek, rozkład staje się bardziej dopadowany do normalnego (równomierne rozproszenie danych, lepsza koncentracja wokół średniej). Dodatkowo odchylenie standardowe jest bardziej zbliżone do pożądanego w przypadku większej liczby obserwacji.
